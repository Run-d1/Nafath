```python
import pandas as pd
```

```python
df = pd.read_excel('data/nafath OpenData.xlsx')

df
```

## Processing

```python
df = df.drop(['name', 'arabicDescription'], axis=1)

df
```

```python
df['value'] = df['value'].replace({'وزارة العدل': 'Ministry of Justice'})

df
```

```python
df = df.rename(columns={'englishDescription': 'description'})

df
```

```python
df_downloads = df.iloc[15:19]

df_downloads
```

```python
df_downloads.loc[:, 'value'] = (df_downloads['value'].str.strip("M")).astype(float)

df_downloads
```

```python
df_downloads.loc[:, 'value'] = (df_downloads['value']*1000000)

df_downloads
```

```python
df.iloc[15:19] = df_downloads.values
```
